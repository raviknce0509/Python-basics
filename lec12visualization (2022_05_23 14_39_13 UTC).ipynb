# Distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt

#load data
df= sns.load_dataset('iris')
df
```

```python
#distplot function create a histogram
sns.distplot( df["sepal_length"] )
plt.show()
```

```python
#the bins parameter specify the number of bins
sns.distplot( df["sepal_length"], bins=20 )

plt.show()
```

```python
sns.distplot( df["sepal_width"], label= 'sepal_width', color = 'r' )
sns.distplot( df["sepal_length"], label= 'sepal_length', color = 'b')
plt.legend();
```

```python
sns.kdeplot(df['sepal_length'], shade=True, color="b")
```

```python
sns.kdeplot(df['sepal_width'], shade=True, color="r")
sns.kdeplot(df['sepal_length'], shade=True, color="b")
```

```python
#using subplot to create multiple histograms together.
f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)
sns.distplot( df["sepal_length"] , color="skyblue", ax=axes[0, 0])
sns.distplot( df["sepal_width"] , color="olive", ax=axes[0, 1])
sns.distplot( df["petal_length"] , color="gold", ax=axes[1, 0])
sns.distplot( df["petal_width"] , color="teal", ax=axes[1, 1])
```

# Heatmap

```python
import pandas as pd
import numpy as np
df= pd.DataFrame(np.random.random((6,6)),columns=["F1","F2","F3","F4","F5","F6"])
print(df)

# annot: If True, write the data value in each cell.
# annot_kws: Keyword arguments for ax.text when annot is True.
sns.heatmap(df, annot=True, annot_kws={"size": 7})
```

```python
#heatmap using different colors:
sns.heatmap(df, cmap="Blues")
```

```python
sns.heatmap(df, cmap="Greens")
```

# Stacked area plot

```python
dfs1t = fmri[(fmri["subject"]=="s1")&
           (fmri["region"]=="parietal")&
           (fmri["event"]=="stim")]
dfs2t = fmri[(fmri["subject"]=="s1")&
           (fmri["region"]=="frontal")&
           (fmri["event"]=="stim")]
dft = pd.merge(dfs1t,dfs2t,on='timepoint')
dft2c = dft[['timepoint','signal_x','signal_y']]

# When stacked is True, each column must be either all positive or negative
dft2c['signal_x'] = dft2c['signal_x']+0.3
dft2c['signal_y'] = dft2c['signal_y']+0.3
dft2sort = dft2c.sort_values(by=['timepoint'])

ax = dft2sort.plot.area(x='timepoint')
```

# Map

```python
import folium
m = folium.Map(location=[41.8781, -87.6298], zoom_start=12)
m
```

```python
# read the data
url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'
state_geo = f'{url}/us-states.json'
state_unemployment = f'{url}/US_Unemployment_Oct2012.csv'
state_data = pd.read_csv(state_unemployment)

m = folium.Map(location=[48, -102], zoom_start=3)

folium.Choropleth(
    geo_data=state_geo,  #This is the geo JSON file
    name='choropleth',
    data=state_data,     #This is from the data file
    columns=['State', 'Unemployment'],
    key_on='feature.id', #key_on specified which feature to use in the geo JSON file
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Unemployment Rate (%)'
).add_to(m)

folium.LayerControl().add_to(m) # Display the map in Jupyternotebook
m
```
